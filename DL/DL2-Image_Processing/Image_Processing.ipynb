### Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('dog.jpg')
```

```python
img_resize = img.resize((200, 200))
img_resize.show()
```

```python
print(img_resize.size)  # this will give the size of resized image
```

```python
img_resize.save('dog_resized.jpg')  # this will save the resized image with new name
```

```python
# displaying the resized image
img_resize = mapimg.imread('dog_resized.jpg')
img_plot = plt.imshow(img_resize)
plt.show()
```

```python
print(img_resize.shape) # this will give the shape of resized image array
```

### Converting the RGB Image to Grayscale

```python
# Importing opencv library
import cv2 # OpenCV library for image processing
```

```python
img = cv2.imread('dog.jpg')
```

```python
img.shape # this will give the shape of image array
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
grayscale_img_show = plt.imshow(grayscale_img, cmap='gray')
plt.show()
```

```python
grayscale_img.shape # this will give the shape of grayscale image array
```

```python
type(grayscale_img)     # this will give the type of grayscale image array      
```

```python
# resize grayscale image
grayscale_img_resize = cv2.resize(grayscale_img, (200, 200))
```

```python
# lod the grayscale resized image to display
grayscale_img_resize_show = plt.imshow(grayscale_img_resize, cmap='gray')
```

```python
# save the grayscale resized image
cv2.imwrite('dog_grayscale_resized.jpg', grayscale_img_resize)
```
